```python
import pandas as pd


path = "C:/Users/rapa/Desktop/빅데이터분석기사실기/"

train = pd.read_csv(path + "train(21).csv")
test = pd.read_csv(path + "test(21).csv")

test.head()
```

```python
# ** 모듈 이름 암기

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
train_d, val_d = train_test_split(train, test_size = 0.2)


train_y = train_d['Survived']
train_x = train_d.drop('Survived', axis = 1)

val_y = val_d['Survived']
val_x = val_d.drop('Survived', axis = 1)

print(train_y.shape, train_x.shape, test.shape)
```

```python
feature = ['Pclass', 'Sex', 'Age', 'Parch', 'Fare']
train_x = pd.get_dummies(train_x[feature])
test_x = pd.get_dummies(test[feature])
```

```python
test_x.shape, train_x.shape
```

```python
model = RandomForestClassifier()

model.fit(train_x, train_y)
```

```python
pred = model.predict(test_x)

pred_train = model.predict(train_x)


result = accuracy_score(pred_train, train_y)
print(result)
```

```python
# 수험번호 .csv 저장하는 코드
ID = test['PassengerId']
print(ID)
```

```python
re_df = pd.DataFrame({"PassengerID": ID, "Survived": pred})

re_df.head()
```
