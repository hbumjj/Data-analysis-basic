```python
import pandas as pd

path = "C:/Users/rapa/Desktop/빅데이터분석기사실기/covid-vaccination-vs-death_ratio.csv"
df = pd.read_csv(path)
print(df.head())
```

```python
# 마지막 접종률을 기준으로 설정
df2 = df.groupby('country').max() # ** 이부분 공부 필요
print(df2)
```

```python
df2 = df2.sort_values(by = 'ratio', ascending = False) # ** 이 부분 암기 필요/ # 오른차순 false
print(df2)
```

```python
df3 = df2[df2['ratio']<=100]
df3
```

```python
# top 10
top = df3.iloc[:10]['ratio'].mean()

# bottom 10
bottom = df3.iloc[-10:]['ratio'].mean()

print(top)
print(bottom)

round(top - bottom, 1)
```
