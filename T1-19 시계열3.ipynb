```python
# 라이브러리 불러오기
import pandas as pd

path = "C:/Users/rapa/Desktop/빅데이터분석기사실기/basic2(t117).csv"

df = pd.read_csv(path)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python

df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day

# event 칼럼 1인경우 sales의 80%

def event_sales(x):
    if x['Events'] == 1:
        x['Sales2'] = x['Sales']*0.8
    else:
        x['Sales2'] = x['Sales']
    return x

# ** df.apply함수에 대해서 알아 두기
df = df.apply(lambda x: event_sales(x), axis = 1) # 1은 row, 0은 컬럼

df.head()
```

```python
# 2022 월별 sales 합계 중 가장 큰 값
max_2022 = 0

for i in range(1, 13):
    a = df[(df['year'] == 2022) & (df['month'] == i)]['Sales2'].sum()
    max_2022 = max(max_2022, a)

# 2023 월별 sales 합계 중 가장 큰 값

max_2023 = 0

for i in range(1, 13):
    a = df[(df['year'] == 2023) & (df['month'] == i)]['Sales2'].sum()
    max_2023 = max(max_2023, a)

print(round(abs(max_2022 - max_2023),0))
```
