올림: np.ceil()
내림: np.floor( # 내림)
버림: np.trunc # 소수점 버리기
반올림: np.round() # 2개 들어감, (숫자, 자리수)()

```python
# floor는 내림, ceil은 올림, truc 버림, 소수점 반올림은 round

df = a['age']

pre = df[df != np.floor(df)]


floor = np.floor(pre).mean()
ceil = np.ceil(pre).mean()
round = np.trunc(pre).mean()
print(floor + ceil + round)
```
