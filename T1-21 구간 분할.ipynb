```python
import pandas as pd

path = "C:/Users/rapa/Desktop/빅데이터분석기사실기/basic1(t121).csv"

df = pd.read_csv(path)

df.head()
```

```python
import numpy as np

df2 = df[(df['age'] > 0) & (round(df['age'],0) == df['age'])]
```

```python
# 정렬하기 
df3 = df2.sort_values(by = ['age'])

print(df3)
```

```python
# **qcut : 동일한 개수로 나누는 명령어임 암기


pd.qcut(df3['age'], q = 3) # 구간 확인


df3['range'] = pd.qcut(df3['age'], q = 3, labels = ['group1', 'group2', 'group3'])

df3.head()
```

```python
# ** 갯수 세는 명령어 암기
# value_counts()
df3['range'].value_counts()


g1 = df3[df3['range'] == 'group1']['age'].median()
g2 = df3[df3['range'] == 'group2']['age'].median()
g3 = df3[df3['range'] == 'group3']['age'].median()

print(g1 + g2 + g3)
```
